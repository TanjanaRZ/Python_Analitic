```python
plt.bar(df['condition'], df['price'])
plt.title('Зависимость стоимости от состояния домов')
plt.xlabel('Состояние')
plt.ylabel('Стоимость домов');
```

Ожидаемо, что дома с наиболее высокой оценкой состояния должны стоить больше. 
Но как показывает анализ данных - это совершенно не так! Оказывается, что дома с наиболее высокой оценкой состояния стоят значительно ниже домов со средней оценкой -3 и оценкой - 4.

```python
plt.figure(figsize=(8,5))
sns.barplot(x=df['floors'], y=df['price'])
plt.title('Зависимость стоимости от этажности дома')
plt.xlabel('Количество этажей')
plt.ylabel('Стоимость домов');
```

**Вывод:** Самые дорогие дома имеют 2,5 этажа, менее дорогие - 3,5, далее 2, 3, 1.5 и 1.

Очень интересно, влияет ли этажность дома на стоимость? Проведя данный анализ можно сделать вывод, что чем выше этаж - тем дороже, но ... только до 2,5 этажа! Выше этажность не так влияет на стоимость.

```python
plt.figure(figsize=(8,5))
sns.barplot(x=df['view'], y=df['price'])
plt.title('Зависимость стоимости от вида')
plt.xlabel('Оценка вида')
plt.ylabel('Стоимость домов');
```

**Вывод:** Стоимость домов напрямую зависит от вида. Чем лучше вид, тем дороже дом.

Данный вывод очень ожидаем! Кто же сомневался, что вид очень влияет на стоимость!

```python
data5 = pd.crosstab(index = df['sqft_living'] ,columns = df['price']).reset_index()
```

```python

plt.figure(figsize=(10,4))
n_ticks = np.arange(len(data5['sqft_living']))
offset = 0.2
w = 0.4
plt.bar(n_ticks-offset,data5['sqft_living'], width=w)
plt.bar(n_ticks-offset,data5['sqft_living'], width=w)
plt.title('Зависимость стоимости то жилой площади')
plt.xlabel('Стоимость')
plt.ylabel('Жилая площадь');
```

**Вывод:** Стоимость домов напрямую зависит от жилой площади. Чем больше жилая площадь, тем дороже дом.

Данный вывод также очень ожидаем! Кто же сомневался, что размер жилой площади очень влияет на стоимость!

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix,1)
corr_matrix[np.abs(corr_matrix)<0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(20,16))
sns.set(font_scale = 1.4)
sns.heatmap(corr_matrix ,annot=True, linewidths=.5,cmap= 'coolwarm')
plt.title('Матрица корреляций');
```

**Вывод:** По данной таблице корреляций можно отследить все существующие зависимости, в том числе и, интересующие нас, зависимости стоимости. Стоимость, судя по таблице не имеет очень высокой зависимости, имеет высокую зависимость от размера жилой площади и оценки Grade (0.7). Немного менее зависима стоимость от наличия чердака и показателя 	sqft_living15 (0,6). Также стоимость имеет умеренную и слабую зависимости с некоторыми признаками.

Анализ с помощью корреляционной таблицы очень интересен, потому что с его помощью мы можем отследить любые зависимости и их силу. И далее продолжить анализ интересующих нас данных более детально.
